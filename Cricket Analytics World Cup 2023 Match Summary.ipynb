```python
from bs4 import BeautifulSoup
import requests
import time
import pandas as pd
import numpy as np
import urllib.request
```

```python
html_text = requests.get('https://www.espncricinfo.com/records/tournament/team-match-results/icc-cricket-world-cup-2023-24-15338').text
```

```python
soup = BeautifulSoup(html_text, 'lxml')
```

```python
my_soup_data = soup.find_all('tr')
```

```python
array_list = [[]]


for matches in my_soup_data: 
    lists = matches.find_all('span', class_= '') 
    temp = []
    for auto in lists: 
        temp.append(auto.text)
    lists2 = matches.find_all('a', class_= 'ds-inline-flex ds-items-start ds-leading-none')
    t = 0
    for link in lists2:
        if t == 0:
            t+=1
            continue
        the_link = link.get("href")
        temp.append(the_link)
    array_list.append(temp)
print(array_list)
team1 = []
team2 = []
winner = []
wonBy = []
location = []
date = []
matchId = []
scoreCard = []

for auto in array_list:
    if len(auto) != 0:
        team1.append(auto[0])
        team2.append(auto[1])
        winner.append(auto[2])
        wonBy.append(auto[3])
        location.append(auto[4])
        date.append(auto[5])
        matchId.append(auto[6])
        scoreCard.append('https://www.espncricinfo.com' + auto[7])
    # print(scoreCard)
```

```python
df = pd.DataFrame(list(zip(team1, team2, winner, wonBy, location, date, matchId, scoreCard ))
    , columns=['Team 1', 'Team 2', 'Winner', 'Won By', 'Location','Date','Match ID', 'Scorecard'])
```

```python
df['Scoreboard'] = df.apply(lambda row: f'=HYPERLINK("{row["Scorecard"]}", "{row["Match ID"]}")', axis=1)
```

```python
df = df.drop(['Scorecard', 'Match ID'], axis=1)
```

```python
players_info = []
players_info_link = []
map_players = {}
Images = []
agePlayers = []
battingStyle = []
bowlingStyle = []
playingRole = []
x = 0
for auto in scoreCard:
    x += 1
    html_text = requests.get(auto).text
    soup = BeautifulSoup(html_text, 'lxml')
    soup_data = soup.find_all('tbody', class_='')
    other_temp_data = soup.find_all('div', class_='ds-text-tight-s ds-font-regular ds-leading-4 ds-text-typo-mid1')


    for auto in soup_data:
        temp_data = auto.find_all('td', class_='ds-w-0 ds-whitespace-nowrap ds-min-w-max ds-flex ds-items-center') + auto.find_all('td' , class_='ds-w-0 ds-whitespace-nowrap ds-min-w-max ds-flex ds-items-center ds-border-line-primary ci-scorecard-player-notout') 
        for juto in temp_data:
            player_name = juto.text
            if player_name[-1] == '':
                player_name = player_name[:-1]
            if player_name[-1] == ')':
                player_name = player_name[:-3]
            
            the_link = 'https://www.espncricinfo.com' + juto.find('a', class_='ds-inline-flex ds-items-start ds-leading-none')['href']
            # print(player_name, " ", 'https://www.espncricinfo.com' + the_link)
            # print(the_link)
            new_html_text = requests.get(the_link).text
            new_soup = BeautifulSoup(new_html_text, 'lxml')
            new_soup_data = new_soup.find_all('div', class_= "ds-grid lg:ds-grid-cols-3 ds-grid-cols-2 ds-gap-4 ds-mb-8")

            for i in range(len(new_soup_data)):
                for itr in new_soup_data[i]:
                    # print(itr.text)
                    temp = itr.text
                    # print(temp+'\n')
                    if 'Age' in temp:
                        agePlayers.append(temp.split('Age')[1].split('y')[0].strip())
                    if 'Batting Style' in temp:
                        battingStyle.append(temp.split('Batting Style')[1].strip())
                    if 'Bowling Style' in temp:
                        bowlingStyle.append(temp.split('Bowling Style')[1].strip()) 
                    if 'Playing Role' in temp:
                        playingRole.append(temp.split('Playing Role')[1].strip()) 
                break

            # print(new_soup_data)
            players_info.append(player_name)
            players_info_link.append(the_link)

            # break

            # print(juto)


    for auto in other_temp_data:
        l = len(auto)
        c = 0
        for buto in auto: 
            # print(buto)
            c+=1
            player_name = str(buto.text)
            if player_name == "Did not bat: ":
                continue
            if c != l:
                player_name = player_name[:-2]
            if player_name[-1] == '':
                player_name = player_name[:-1]
            if player_name[-1] == ')':
                player_name = player_name[:-3]
            for link in buto:
                the_link = 'https://www.espncricinfo.com' + link.get("href")

            new_html_text = requests.get(the_link).text
            new_soup = BeautifulSoup(new_html_text, 'lxml')
            new_soup_data = new_soup.find_all('div', class_= "ds-grid lg:ds-grid-cols-3 ds-grid-cols-2 ds-gap-4 ds-mb-8")

            for i in range(len(new_soup_data)):
                for itr in new_soup_data[i]:
                    # print(itr.text)
                    temp = itr.text
                    # print(temp+'\n')
                    if 'Age' in temp:
                        agePlayers.append(temp.split('Age')[1].split('y')[0].strip())
                    if 'Batting Style' in temp:
                        battingStyle.append(temp.split('Batting Style')[1].strip())
                    if 'Bowling Style' in temp:
                        bowlingStyle.append(temp.split('Bowling Style')[1].strip()) 
                    if 'Playing Role' in temp:
                        playingRole.append(temp.split('Playing Role')[1].strip()) 
                break

            players_info.append(player_name)
            players_info_link.append(the_link)
```

```python
df2 = pd.DataFrame(list(zip(players_info, players_info_link, agePlayers, battingStyle, bowlingStyle, playingRole))
    , columns=['Player Name','Information', 'Players Age', 'Batting Style','Bowling Style','Playing Role'])
df2.head()
```

```python
df2['Player_Name'] = df2.apply(lambda row: f'=HYPERLINK("{row["Information"]}", "{row["Player Name"]}")', axis=1)
df2 = df2.drop(['Player Name', 'Information'], axis = 1)
columns = list(df2.columns)
columns = [columns[-1]] + columns[:-1]
df2 = df2[columns]
df2 
```
